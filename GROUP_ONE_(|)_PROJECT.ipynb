```python
# Register students and their courses with an initial grade of 0
def create_grades_dataframe(students):
    """
    Function to create a DataFrame of students with their Trainee IDs and initial course grades.

    Parameters:
    students (dict): Dictionary of student data, including Trainee IDs and course grades.

    Returns:
    DataFrame: A pandas DataFrame containing student data.
    """
    courses = list(next(iter(students.values()))["Courses"].keys())  # Extract courses dynamically
    data = {
        "Student": [],
        "Trainee ID": [],
    }
    data.update({course: [] for course in courses})  # Add courses to the data dictionary

    for student, details in students.items():
        data["Student"].append(student)
        data["Trainee ID"].append(details["Trainee ID"])
        for course in courses:
            data[course].append(details["Courses"].get(course, 0))

    return pd.DataFrame(data)

# Creating the DataFrame
grades_df = create_grades_dataframe(students_data)
grades_df
```

```python
# Add/Update multiple grades for a student
def add_or_update_multiple_grades(df, student_name, courses_and_new_grades):
    """
    Function to add or update multiple grades for a student in the DataFrame.

    Parameters:
    df (DataFrame): The DataFrame containing student grades.
    student_name (str): The name of the student whose grades are to be updated.
    courses_and_new_grades (list of tuples): List of tuples with (course, new_grade).
    """
    if student_name not in df["Student"].values:
        print(f"Error: Student '{student_name}' not found in the DataFrame.")
        return

    for course, new_grade in courses_and_new_grades:
        if course not in df.columns:
            print(f"Warning: Course '{course}' does not exist in the DataFrame. Skipping update for {student_name}.")
            continue
        df.loc[df["Student"] == student_name, course] = new_grade
        print(f"Updated {student_name}'s grade for {course} to {new_grade}.")
```

```python
import pandas as pd

# DataFrame containing student grades
data = {
    "Student": [
        "Adeniran", "Iretomiwa", "Samson", "Bala", "Bonaventure",
        "Christian", "Seyi", "Iyokho", "Kolawole", "Okanlawon"
    ],
    "Python Programming": [68, 78, 46, 91, 49, 65, 60, 72, 52, 42],
    "Data Science": [64, 47, 84, 57, 97, 51, 64, 100, 38, 69],
    "Statistics": [66, 85, 49, 49, 64, 38, 49, 85, 77, 62],
    "Mathematics": [73, 78, 94, 60, 66, 90, 100, 93, 83, 63],
    "Machine Learning": [97, 95, 67, 75, 77, 40, 54, 82, 95, 50]
}

grades_df = pd.DataFrame(data)

# Function to find the highest score and the respective students for multiple courses
def get_highest_scores(df, courses):
    """
    Function to find and display the highest scores along with the students
    who achieved them for a given list of courses.

    Parameters:
    df (DataFrame): The DataFrame containing grades.
    courses (list): List of course names for which to find the highest scores.

    Returns:
    None: Prints the highest scores and corresponding students.
    """
    for course in courses:
        # Validate if the course exists in the DataFrame
        if course not in df.columns:
            print(f"Error: Course '{course}' does not exist in the DataFrame.")
            continue

        # Find the maximum score for the course
        max_score = df[course].max()
        # Find the students who achieved the maximum score
        students = df[df[course] == max_score]["Student"].values
        print(f"The highest score in {course} is {max_score}, achieved by: {', '.join(students)}.")

# List of courses
courses = [
    "Python Programming",
    "Data Science",
    "Statistics",
    "Mathematics",
    "Machine Learning"
]

# Function call to display the highest scores
get_highest_scores(grades_df, courses)
```
